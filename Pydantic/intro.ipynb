Model with Optional Fields

Add optional fields using python's optional type

```python
from typing import Optional

class Employee(BaseModel):
    id: int
    name: str
    department: str
    salary: Optional[float] = None # Optional with default value
    is_active: Optional[float] = True # Optional with default value
```

```python
employee = Employee(id=101,name="Sai",department="AI")
employee
```

Create comple strctures with nested models

```python
from pydantic import BaseModel

class Address(BaseModel):
    street: str
    city: str
    zip_code: int

class Customer(BaseModel):
    customer_id: int
    name: str
    address: Address # Nested model


customer = Customer(
    customer_id = 1,
    name = "Sai",
    address = {"street": "123 Friends colony", "city": "Hyderabad","zip_code":"2108"}
)
print(customer)
```

In Pydantic, type coercion happens automatically for compatible values. In my example, the zip_code field is defined as an int, but I passed "2108" as a string inside the nested Address model while creating the Customer object. Pydantic accepted this because the string "2108" is a numeric string and can be safely converted using int("2108"), so it automatically cast it to the integer 2108 without raising an error. However, when the field type is defined as str and an integer like 2108 is provided instead, Pydantic (especially in v2 or strict configurations) may raise a validation error because it does not always implicitly convert integers to strings. This stricter behavior is intentional to prevent unintended data transformations. Therefore, string-to-integer conversion is generally allowed when the string clearly represents a number, but integer-to-string conversion can be restricted depending on the model configuration and Pydantic version.

Pydantic fields: Customisation and Constraints

Pydantic's Field validation:

ge = greater than or equal to (≥)

gt = greater than (>)

le = less than or equal to (≤)

lt = less than (<)

```python
from pydantic import BaseModel, Field

class Item(BaseModel):
    name:str= Field(min_length=2, max_length=50)
    price:float=Field(ge=0,le=1000)
    quantity:int=Field(ge=0)
item = Item(name="Art", price=0,quantity=0)
print(item)
```

```python
from pydantic import BaseModel, Field

class User(BaseModel):
    username:str= Field(...,description="The unique username of the user")
    age:float=Field(default=18,description="The age of the user, defaults to 18")
    email:int=Field(default_factory=lambda: "user@example.com",description="default email address")
user1 = User(username="Sai")
print(user1)
```
